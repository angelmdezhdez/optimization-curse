## Ejercicio 1 (2.5 puntos)

1. Encuentre y clasifique los puntos estacionarios para la función

$$ f(\mathbf{x}) = x_1^2 - x_2^2 + x_3^2 - 2x_1x_3 - x_2x_3 + 4x_1 + 12. $$

2. Sea $\mathbf{x}_0 = (1,0,0)^\top$. Calcule el punto $\mathbf{x}_1$ 
   usando la dirección de descenso máximo con paso exacto.


### Solución:

1. Calculemos las derivadas parciales de $f$, estas son:  $$\dfrac{\partial f}{\partial x_1}=2x_1 -2x_3 +4$$ $$\dfrac{\partial f}{\partial x_2}=-2x_2-x_3$$ $$\dfrac{\partial f}{\partial x_3}=2x_3 -2x_1 -x_2$$Si igualamos cada derivada a 0 y resolvemos el sistema de ecuaciones resultante obtenemos que el único punto estacionario es: $$\mathbf{x}=\begin{pmatrix} -10 \\ 4 \\ -8 \end{pmatrix}$$

Calculando la matriz Hessiana de $f$ obtenemos $$H= \begin{pmatrix} 2 & 0 &-2 \\ 0 & -2 & -1 \\ -2 & -1 & 2\end{pmatrix}$$ que tiene como eigenvalores $\lambda_1\approx 0.21, \lambda_2\approx -2.29, \lambda_3\approx 4.08$ por lo que el punto es de inflexión,

2. Tenemos que $$\mathbf{p}_0=\begin{pmatrix} -6 \\ 0 \\ 2\end{pmatrix}$$ por lo que $$\mathbf{x}_0 + \alpha\mathbf{p}_0 = \begin{pmatrix} 1-6\alpha \\ 0 \\ 2\alpha\end{pmatrix}$$ entonces $$f(\mathbf{x}_0 + \alpha\mathbf{p}_0) = 64\alpha^2 -40\alpha +17 \Rightarrow f'(\mathbf{x}_0 + \alpha\mathbf{p}_0)= 128\alpha -40$$ por lo que, el valor de $\alpha$ que minimice la expresión es $\alpha =\dfrac{5}{16}$ y así $$\mathbf{x}_1=\begin{pmatrix} -\dfrac{7}{8} \\ 0 \\ \dfrac{5}{8}\end{pmatrix}$$

## Ejercicio 3 (2 puntos)

Supongamos que $f_1, f_2: \mathbb{R}^n \rightarrow \mathbb{R}$ son funciones convexas.
   
1. Muestre que también es convexa la función $f(\mathbf{x})$ 
   definida como
   
$$ f(\mathbf{x}) = \max\{f_1(\mathbf{x}), f_2(\mathbf{x}) \}. $$

2. Si $n=1$ y $f_1(-0.4) = 0.36$,
   $f_1(0.6) = 2.56$, $f_2(-0.4) = 3.66$ y $f_2(1) = 2$, identifique 
   el intervalo más pequeño en el que se puede garantizar que se 
   encuentra el minimizador de la función $f(x)$. 
   Explique su respuesta.

### Solución:

1. Sean $f_1$ y $f_2$ funciones convexas en $C\subset\mathbb{R}^n$ y sea $f(x)=\max\{f_1(x),f_2(x)\}$.
Tomemos $x,y\in C$ y $\lambda\in [0,1]$, y sea $z=\lambda x +(1-\lambda )y$, sin pérdida de generalidad, supongamos que $$\begin{split} 
f(z) &= f_1(z) \\
&\leq \lambda f_1(x) +(1-\lambda)f_1(y) \\
&\leq \lambda f(x) +(1-\lambda)f(y)
\end{split}$$
con lo que se demuestra que $f$ es convexa.
2. Consideremos las rectas $l_1$ y $l_2$ que conectan los dos puntos dados de cada función, entonces: $$l_1(x)=2.2(x+0.4)+0.36$$ $$l_2(x)= -\dfrac{1.66}{1.4}(x+0.4)+3.66$$
Estas rectas tienen pendientes de signo distinto y están definidas en el mismo intervalo, por lo que existe una intersección entre estas rectas en el intervalo $[-0.4,0.6]$. Es fácil ver que $$f(x)\leq \max\{l_1(x),l_2(x)\} =g(x)$$ para toda $x\in [-0.4, 1]$ y además son iguales en los extremos del intervalo, por construcción. Además, $g$ tiene un mínimo, que es el punto de intersección de ambas rectas, por lo que la función $f$ está dominada por $g$, así que, como $g$ tiene un mínimizador en $[-0.4,0.6]$, $f$ también lo tiene en el mismo intervalo.

````python
#importamos librerías
import numpy as np
import matplotlib.pyplot as plt
````

````python
#definimos las funciones para las hessianas
def Hes_himmelbleau(x):
  d1=12*(x[0]**2) +4*x[1] -42
  d2=4*(x[0]+x[1])
  d3=4*x[0] +12*(x[1]**2) -26
  return np.array([[d1, d2], [d2, d3]])

def Hes_beale(x):
  d1=2*(x[1]**6 +x[1]**4 -2*x[1]**3 -x[1]**2 -2*x[1] +3)
  d2=4*x[0]*(3*(x[1]**5) +2*(x[1]**3) -3*(x[1]**2) -x[1] -1) +15.75*(x[1]**2) +9*x[1] +3
  d3=6*x[0]*(5*x[0]*(x[1]**4 +0.4*(x[1]**2) -0.4*x[1] -0.066666) +5.25*x[1] +1.5)
  return np.array([[d1, d2], [d2, d3]])

def Hes_rosenbrook(x):
  n=len(x)
  m=np.zeros([n,n])
  m[0][0]=1200*(x[0])**2 - 400*(x[1]) +2
  m[n-1][n-1]=200
  m[0][1]=m[1][0]=-400*x[0]
  m[n-1][n-2]=m[n-2][n-1]=-400*x[n-2]
  for i in range(1,n-1):
    m[i][i]=200*(6*(x[i]**2)-2*(x[i+1])+1)+2
    m[i][i-1]=m[i-1][i]=-400*(x[i-1])
    m[i][i+1]=m[i+1][i]=-400*(x[i])
  return m

alfa = np.array([1.0, 1.2, 3.0, 3.2])

A = np.array([[10, 3, 17, 3.5, 1.7, 8],
              [0.05, 10, 17, 0.1, 8, 14],
                [3, 3.5, 1.7, 10, 17, 8],
                [17, 8, 0.05, 10, 0.1, 14]])


P = 10**(-4) * np.array([[1312, 1696, 5569, 124, 8283, 5886],
                          [2329, 4135, 8307, 3736, 1004, 9991],
                          [2348, 1451, 3522, 2883, 3047, 6650],
                          [4047, 8828, 8732, 5743, 1091, 381]])

def hartman(x):
  sum1=0
  for k in range(4):
    sum2=0
    for j in range(6):
      sum2+=A[k][j]*((x[j]-P[k][j])**2)
    sum1+= alfa[k]*np.exp(-sum2)/1.94
  return -(1/1.94)*(2.58 + sum1)

def Grad_hartman(x):
  g=np.zeros(6)
  for i in range(6):
    sum1=0
    for k in range(4):
      sum2=0
      for j in range(6):
        sum2+=A[k][j]*((x[j]-P[k][j])**2)
      sum1+= 2*alfa[k]*A[k][i]*(x[i] - P[k][i])*np.exp(-sum2)/1.94
    g[i]=sum1
  return g

def Hes_hartman(x):
  H=np.zeros([6,6])
  for i in range(6):
    for l in range(i, 6):
      sum1=0
      for k in range(4):
        sum2=0
        for j in range(6):
          sum2+=A[k][j]*((x[j]-P[k][j])**2)
        sum1+=2*alfa[k]*A[k][i]*np.exp(-sum2)*(1-2*A[k][l]*(x[l]-P[k][l])*(x[i]-P[k][i]))/1.94
      H[i][l]=H[l][i]=sum1
  return H

def himmelblau(x):
  return (x[0]**2+x[1]-11)**2 + (x[0]+x[1]**2-7)**2

def Grad_himmelblau(x):
  d1=4*x[0]**3 + 4*x[0]*x[1] - 42*x[0] + 2*x[1]**2 - 14
  d2=4*x[1]**3 + 4*x[0]*x[1] + 2*x[0]**2 - 26*x[1] - 22
  return np.array([d1,d2])

def beale(x):
  return (1.5-x[0]+x[1]*x[0])**2 + (2.25-x[0]+x[0]*x[1]**2)**2 + (2.625-x[0]+x[0]*x[1]**3)**2

def Grad_beale(x):
  d1=2*(x[1]-1)*(1.5-x[0]+x[1]*x[0])+2*(x[1]**2-1)*(2.25-x[0]+x[0]*x[1]**2)+2*(x[1]**3-1)*(2.625-x[0]+x[0]*x[1]**3)
  d2=2*(x[0])*(1.5-x[0]+x[1]*x[0])+4*(x[0]*x[1])*(2.25-x[0]+x[0]*x[1]**2)+6*(x[0]*x[1]**2)*(2.625-x[0]+x[0]*x[1]**3)
  return np.array([d1, d2])

def rosenbrock(x):
  f=0
  n=x.shape[0]
  for i in range(n-1):
    f+=100*(x[i+1]-x[i]**2)**2 + (1-x[i])**2
  return f

def Grad_rosenbrock(x):
  n=x.shape[0]
  grad=np.zeros(n)
  grad[0]=-2*(1-x[0])-400*x[0]*(x[1]-x[0]**2)
  grad[-1]=200*(x[-1]-x[-2]**2)
  for i in range(1,n-1):
    grad[i]=200*(x[i]-x[i-1]**2-2*x[i]*x[i+1]+2*x[i]**3) -2*(1-x[i])
  return grad
````

````python
def contornosFnc2D_modified(fncf, xleft, xright, ybottom, ytop, levels, puntos):
    # Crear una discretización uniforme del intervalo [xleft, xright]
    ax = np.linspace(xleft, xright, 250)
    # Crear una discretización uniforme del intervalo [ybottom, ytop]
    ay = np.linspace(ybottom, ytop, 200)
    # La matriz mX que tiene las abscisas
    mX, mY = np.meshgrid(ax, ay)
    # Se crea el arreglo mZ con los valores de la función en cada nodo
    mZ = mX.copy()
    for i, y in enumerate(ay):
        for j, x in enumerate(ax):
            mZ[i, j] = fncf(np.array([x, y]))

    # Grafica de las curvas de nivel
    fig, ax = plt.subplots()
    CS = ax.contour(mX, mY, mZ, levels, cmap='Wistia')

    puntos = np.array(puntos)
    ax.plot(puntos[:, 0], puntos[:, 1], color='red', label='Puntos')

    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Curvas de Nivel y Puntos')
    plt.legend()
    plt.show()
````

````python
def solution(sol, f, df):
  print("El minimizador alcanzado es {}".format(sol[0]))
  print("¿Terminó por criterio de paro? {}".format(sol[1]))
  print("Iteraciones realizadas: {}".format(sol[2]))
  print("f(xk)={}".format(f(sol[0])))
  print("||Df(xk)||={}".format(np.linalg.norm(df(sol[0]))))
````

````python
eps=np.finfo(float).eps
````

````python
# programamos la aproximacion cuadratica
def mk(f, grad, hess, xk, p):
    return f(xk) + p.T @ grad(xk) + (p.T @ hess(xk) @ p)/2
````

````python
# Programamos la función
def MRC(f, grad, hess, x0, maxIter, tol, rmax, rmin, n, eps):
    xk = x0
    seqx = [xk.copy()]
    r = (rmin + rmax)/4.0
    dim = len(xk)
    for k in range(maxIter):
        gk = grad(xk)
        gk_norm = np.linalg.norm(gk)
        if gk_norm < tol:
            if dim == 2:
                return xk, True, k+1, seqx
            else:
                return xk, True, k+1, None
        Bk = hess(xk)
        z = gk.T @ Bk @ gk
        if z <= eps:
            tk = 1.0
        else:
            tk = min(1.0, (gk_norm**3)/(r*z))
        pkc = -tk*r*gk/gk_norm
        rhok = (f(xk) - f(xk + pkc))/(f(xk) - mk(f, grad, hess, xk, pkc))
        if rhok < 0.25 and r > 4*rmin:
            r = 0.25 * r
        elif rhok > 0.75 and np.linalg.norm(pkc) == r:
            r = min(rmax, 2*r)
        else: 
            r = r
        if rhok > n:
            xk = xk + pkc
        else:
            xk = xk
        seqx.append(xk.copy())
    if dim == 2:
        return xk, False, maxIter, seqx
    else:
        return xk, False, maxIter, None
````

````python
N = 50000
rmin = 10e-5
n = 0.25
tol1 = np.sqrt(2*eps)
````
