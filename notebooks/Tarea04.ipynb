### Solución:

# Para Himmelbleau:
$$H(f)=\begin{pmatrix}
12x_1^2+4x_2-42 & 4(x_1+x_2) \\
4(x_1+x_2) & 4x_1+12x_2^2-26

# Para Beale:
$$ H(f)= \begin{pmatrix}

# Para Rosenbrook:
Aquí expresaremos las derivadas parciales generales, pues recordemos que esta función es generalizable a muchas dimensiones.

\begin{equation*}
\dfrac{\partial ^2 f}{\partial x_1^2} = 1200x_1^2-400x_2+2
\end{equation*}

Para $i\geq 2$:
\begin{equation*}
\dfrac{\partial ^2 f}{\partial x_i^2} = 200(6x_i^2 -2x_{i+1}+1)+2
\end{equation*}

\begin{equation*}
\dfrac{\partial ^2 f}{\partial x_i\partial x_{i-1}}= \dfrac{\partial ^2 f}{\partial x_{i-1}\partial x_{i}} = -400x_{i-1}
\end{equation*}

\begin{equation*}
\dfrac{\partial ^2 f}{\partial x_i\partial x_{i+1}}= \dfrac{\partial ^2 f}{\partial x_{i+1}\partial x_{i}} = -400x_{i}
\end{equation*}

\begin{equation*}
\dfrac{\partial ^2 f}{\partial x_n^2} = 200
\end{equation*}

# Para Hartman:
Para las derivadas parciales de primer orden se tiene que
\begin{equation*}
\dfrac{\partial f}{\partial x_i} = \dfrac{2}{1.94}\left[ \displaystyle\sum_{k=1}^4 \alpha_ka_{ki}(x_i-P_{ki})\exp\left( -\sum_{j=1}^6 a_{kj}(x_j-P_{kj})^2 \right) \right]
\end{equation*}

Para las derivadas de segundo orden:
\begin{equation*}
\dfrac{\partial ^2 f}{\partial x_i\partial x_l}=\dfrac{\partial ^2 f}{\partial x_l\partial x_i}=\dfrac{2}{1.94}\left[ \displaystyle\sum_{k=1}^4 \alpha_ka_{ki}\exp\left( -\sum_{j=1}^6 a_{kj}(x_j-P_{kj})^2 \right)\left[ 1-2a_{kl}(x_l-P_{kl})(x_i-P_{ki}) \right] \right]
\end{equation*}

````python
#definimos las funciones para las hessianas
def Hes_himmelbleau(x):
  d1=12*(x[0]**2) +4*x[1] -42
  d2=4*(x[0]+x[1])
  d3=4*x[0] +12*(x[1]**2) -25
  return np.array([[d1, d2], [d2, d3]])

def Hes_beale(x):
  d1=2*(x[1]**6 +x[1]**4 -2*x[1]**3 -x[1]**2 -2*x[1] +3)
  d2=4*x[0]*(3*(x[1]**5) +2*(x[1]**3) -3*(x[1]**2) -x[1] -1) +15.75*(x[1]**2) +9*x[1] +3
  d3=6*x[0]*(5*x[0]*(x[1]**4 +0.4*(x[1]**2) -0.4*x[1] -0.066666) +5.25*x[1] +1.5)
  return np.array([[d1, d2], [d2, d3]])

def Hes_rosenbrook(x):
  n=len(x)
  m=np.zeros([n,n])
  m[0][0]=1200*(x[0])**2 - 400*(x[1]) +2
  m[n-1][n-1]=200
  m[0][1]=m[1][0]=-400*x[0]
  m[n-1][n-2]=m[n-2][n-1]=-400*x[n-2]
  for i in range(1,n-1):
    m[i][i]=200*(6*(x[i]**2)-2*(x[i+1])+1)+2
    m[i][i-1]=m[i-1][i]=-400*(x[i-1])
    m[i][i+1]=m[i+1][i]=-400*(x[i])
  return m

alfa = np.array([1.0, 1.2, 3.0, 3.2])

A = np.array([[10, 3, 17, 3.5, 1.7, 8],
              [0.05, 10, 17, 0.1, 8, 14],
                [3, 3.5, 1.7, 10, 17, 8],
                [17, 8, 0.05, 10, 0.1, 14]])


P = 10**(-4) * np.array([[1312, 1696, 5569, 124, 8283, 5886],
                          [2329, 4135, 8307, 3736, 1004, 9991],
                          [2348, 1451, 3522, 2883, 3047, 6650],
                          [4047, 8828, 8732, 5743, 1091, 381]])

def hartman(x):
  sum1=0
  for k in range(4):
    sum2=0
    for j in range(6):
      sum2+=A[k][j]*((x[j]-P[k][j])**2)
    sum1+= alfa[k]*np.exp(-sum2)/1.94
  return -(1/1.94)*(2.58 + sum1)

def Grad_hartman(x):
  g=np.zeros(6)
  for i in range(6):
    sum1=0
    for k in range(4):
      sum2=0
      for j in range(6):
        sum2+=A[k][j]*((x[j]-P[k][j])**2)
      sum1+= 2*alfa[k]*A[k][i]*(x[i] - P[k][i])*np.exp(-sum2)/1.94
    g[i]=sum1
  return g

def Hes_hartman(x):
  H=np.zeros([6,6])
  for i in range(6):
    for l in range(i, 6):
      sum1=0
      for k in range(4):
        sum2=0
        for j in range(6):
          sum2+=A[k][j]*((x[j]-P[k][j])**2)
        sum1+=2*alfa[k]*A[k][i]*np.exp(-sum2)*(1-2*A[k][l]*(x[l]-P[k][l])*(x[i]-P[k][i]))/1.94
      H[i][l]=H[l][i]=sum1
  return H

def himmelblau(x):
  return (x[0]**2+x[1]-11)**2 + (x[0]+x[1]**2-7)**2

def Grad_himmelblau(x):
  d1=4*x[0]**3 + 4*x[0]*x[1] - 42*x[0] + 2*x[1]**2 - 14
  d2=4*x[1]**3 + 4*x[0]*x[1] + 2*x[0]**2 - 26*x[1] - 22
  return np.array([d1,d2])

def beale(x):
  return (1.5-x[0]+x[1]*x[0])**2 + (2.25-x[0]+x[0]*x[1]**2)**2 + (2.625-x[0]+x[0]*x[1]**3)**2

def Grad_beale(x):
  d1=2*(x[1]-1)*(1.5-x[0]+x[1]*x[0])+2*(x[1]**2-1)*(2.25-x[0]+x[0]*x[1]**2)+2*(x[1]**3-1)*(2.625-x[0]+x[0]*x[1]**3)
  d2=2*(x[0])*(1.5-x[0]+x[1]*x[0])+4*(x[0]*x[1])*(2.25-x[0]+x[0]*x[1]**2)+6*(x[0]*x[1]**2)*(2.625-x[0]+x[0]*x[1]**3)
  return np.array([d1, d2])

def rosenbrock(x):
  f=0
  n=x.shape[0]
  for i in range(n-1):
    f+=100*(x[i+1]-x[i]**2)**2 + (1-x[i])**2
  return f

def Grad_rosenbrock(x):
  n=x.shape[0]
  grad=np.zeros(n)
  grad[0]=-2*(1-x[0])-400*x[0]*(x[1]-x[0]**2)
  grad[-1]=200*(x[-1]-x[-2]**2)
  for i in range(1,n-1):
    grad[i]=200*(x[i]-x[i-1]**2-2*x[i]*x[i+1]+2*x[i]**3) -2*(1-x[i])
  return grad
````

## Ejercicio 2 (2 puntos)

Programe el método de Newton "puro" (Algoritmo 1 de la Clase 10).

1. La función que implementa el algoritmo
   calcula la dirección de Newton resolviendo el sistema
   de ecuaciones $\mathbf{H}_k \mathbf{p}_k^N = - \mathbf{g_k}$.
   No use la factorización de Cholesky porque no hay garantía

   ver como se comporta el algoritmo.
- Si la dimensión $n$ de la variable $\mathbf{x}$ es 2, defina un
  arreglo $\mathbf{M}$  en el que se guarde los puntos $\mathbf{x}_0, \mathbf{x}_1,
  ..., \mathbf{x}_k$ generados. Si $n>2$, devolve $\mathbf{M}$ como $None$.
- La función debe devolver el último valor $k$, $\mathbf{x}_k$,

2. Pruebe el algoritmo usando la cantidad de iteraciones máximas $N=1000$ y la tolerancia

  ..., \mathbf{x}_k$.
  
  
### Solución:

````python
#creamos una funcion para resolver el sistema
def solve_system(A, b):
    try: #para resolver el sistema
        x=np.linalg.solve(A, b)
        return x
    except np.linalg.LinAlgError:
        #en caso de que no se pueda resolver
        return None
````

````python
#creamos la función para metodo de Newton

def Newton_Method(grad, hess, x0, tol, maxIter):
  xk=x0
  n=len(xk)
  seqx=[xk.copy()]
  for k in range(maxIter):
    gk=grad(xk)
    if np.linalg.norm(gk)<tol:
      if n==2:
        return xk, True, k+1, seqx
      else:
        return xk, True, k+1, None
    Hk=hess(xk)
    pk=solve_system(Hk, -gk)
    xk=xk+pk
    seqx.append(xk.copy())
  if n==2:
    return xk, False, maxIter, seqx
  else:
    return xk, False, maxIter, None
````

````python
def contornosFnc2D_modified(fncf, xleft, xright, ybottom, ytop, levels, puntos):
    # Crear una discretización uniforme del intervalo [xleft, xright]
    ax = np.linspace(xleft, xright, 250)
    # Crear una discretización uniforme del intervalo [ybottom, ytop]
    ay = np.linspace(ybottom, ytop, 200)
    # La matriz mX que tiene las abscisas
    mX, mY = np.meshgrid(ax, ay)
    # Se crea el arreglo mZ con los valores de la función en cada nodo
    mZ = mX.copy()
    for i, y in enumerate(ay):
        for j, x in enumerate(ax):
            mZ[i, j] = fncf(np.array([x, y]))

    # Grafica de las curvas de nivel
    fig, ax = plt.subplots()
    CS = ax.contour(mX, mY, mZ, levels, cmap='Wistia')

    puntos = np.array(puntos)
    ax.plot(puntos[:, 0], puntos[:, 1], color='red', label='Puntos')

    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Curvas de Nivel y Puntos')
    plt.legend()
    plt.show()
````

````python
def solution(sol, f, df):
  print("El minimizador alcanzado es {}".format(sol[0]))
  print("¿Terminó por criterio de paro? {}".format(sol[1]))
  print("Iteraciones realizadas: {}".format(sol[2]))
  print("f(xk)={}".format(f(sol[0])))
  print("||Df(xk)||={}".format(np.linalg.norm(df(sol[0]))))
````

## Ejercicio 3 (3 puntos)

Modifique la función que implementa el método de Newton del Ejercicio 1
para incluir el cálculo del tamaño de paso (Newton amortiguado).

1. Además de los parámetros que se mencionan en el Algoritmo 1, la
   función que implementa el método debe recibir los parámetros
   del algoritmo de backtracking: $\rho>0$, $c_1>0$ y el número de
   iteraciones máximas $N_b$. Fijamos el valor inicial $\alpha_{ini}=1$
   para el algoritmo de backtracking para intentar dar el paso completo
   como lo hace el método de Newton, pero
   si ese paso no satisface la condición de descenso, dejamos que lo recorte.
   
   No tenemos garantizado que la matriz Hessiana $\mathbf{H}_k$ sea
   definida positiva, así que agregamos los siguientes pasos al algoritmo
   después de calcular la dirección $\mathbf{p}_k$:
- Si $\mathbf{g}_k^\top\mathbf{p}_k>0$, hacer $\mathbf{p}_k = -\mathbf{p}_k$
  para que sea dirección de descenso.
- Calcular $\alpha_k$ usando el algoritmo de backtracking.
- Calcular $\mathbf{x}_{k+1} = \mathbf{x}_{k} + \alpha_k \mathbf{p}_{k}$.
   

2. Pruebe el algoritmo usando los puntos iniciales del Ejercicio 1,
   fijando la cantidad de iteraciones máximas $N=1000$, la tolerancia
   $\tau = \sqrt{n\epsilon_m}$, donde $\epsilon_m$ es el épsilon de máquina.
   Para el algoritmo de backtracking use $\rho=0.5, c_1=0.1$ y $N_b=500$.
   
### Solución:

````python
#traemos la funcion de bt
def backtracking(a0, r, c, xk, f, df, pk, maxIter):
  a=a0
  for i in range(maxIter):
    if f(xk+a*pk)<= f(xk) + c*a*(np.dot(df(xk), pk)):
      return a, i+1, True
    a=r*a
  return a, maxIter, False
````

````python
#modificamos la función

def Newton_Method_modified(f, grad, hess, x0, tol, maxIter, a0, r, c, maxIter2, tol2):
  xk=x0
  ak=a0
  n=len(xk)
  seqx=[xk.copy()]
  for k in range(maxIter):
    gk=grad(xk)
    if np.linalg.norm(gk)<tol:
      if n==2:
        return xk, True, k+1, seqx
      else:
        return xk, True, k+1, None
    Hk=hess(xk)
    pk=solve_system(Hk, -gk)
    if np.dot(pk, gk)>0:
      pk=-pk
    z=backtracking(ak, r, c, xk, f, grad, pk, maxIter2)
    if z[-1]==False:
      print("BT esta consumiendo todas las iteraciones")
    ak=z[0]
    xk=xk+ak*pk
    seqx.append(xk.copy())
  if n==2:
    return xk, False, maxIter, seqx
  else:
    return xk, False, maxIter, None
````

````python
def Newton_Method_hessian_modified(f, grad, hess, x0, tol, maxIter, a0, r, c, maxIter2, tol2, d):
  xk=x0
  ak=a0
  n=len(xk)
  seqx=[xk.copy()]
  for k in range(maxIter):
    gk=grad(xk)
    if np.linalg.norm(gk)<tol:
      if n==2:
        return xk, True, k+1, seqx
      else:
        return xk, True, k+1, None
    Hk=hess(xk)
    #obtenemos los valores y vectores propios
    values, vectors = np.linalg.eig(Hk)
    #modificamos los valores
    values=np.where(values<d, d, values)
    #reconstruimos la matriz
    Hk=vectors@np.diag(values)@vectors.T
    pk=solve_system(Hk, -gk)
    if np.dot(pk, gk)>0:
      pk=-pk
    z=backtracking(ak, r, c, xk, f, grad, pk, maxIter2)
    if z[-1]==False:
      print("BT esta consumiendo todas las iteraciones")
    ak=z[0]
    xk=xk+ak*pk
    seqx.append(xk.copy())
  if n==2:
    return xk, False, maxIter, seqx
  else:
    return xk, False, maxIter, None
````
